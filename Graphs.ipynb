```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
G.add_node('Alice')
G.add_node("Bob")
G.add_node("Chuck")
```

```python
#View the nodes
G.nodes()
```

```python
#Adding edges to nodes
G.add_edge('Alice','Bob')
G.add_edge('Alice','Chuck')
G.add_edge('Bob','Alice')
G.add_edge('Bob','Chuck')
```

```python
list(G.edges)
```

```python
# Drawing the graph with circles
nx.draw_circular(G, node_color = 'C0', node_size = 2000, with_labels = True)
```

```python
positions = dict(Albany=(-74,43),Boston=(-71,42),NYC=(-74,41),Philly = (-75,40))
G = nx.Graph()
```

```python
G.add_nodes_from(positions)
```

```python
G.nodes()
```

```python
drive_times = {('Albany','Boston'):3,
                ('Albany','NYC'):4,
                ('NYC','Philly'):2}
```

```python
G.add_edges_from(drive_times)
```

```python
nx.draw(G,positions,node_color = 'C1',node_shape = 's',node_size = 2500,with_labels = True)
nx.draw_networkx_edge_labels(G,positions,edge_labels=drive_times)
```
